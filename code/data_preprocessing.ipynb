```python
import pandas as pd
sql_records_df = pd.read_csv("sql_df.csv")
```

```python
sql_records_df
```

```python
from urllib.parse import unquote, urlparse, parse_qs

def url_decoder(full_url):
    full_url = unquote(full_url)
    return full_url

sql_records_df['parsed_url'] = sql_records_df['full_url'].apply(url_decoder)
```

```python
sql_records_df
```

```python
import requests
from bs4 import BeautifulSoup

# scraping keywordow
sql_keywords = []

url = "https://www.w3schools.com/sql/sql_ref_keywords.asp"
resp = requests.get(url)
print(resp.status_code)

soup = BeautifulSoup(resp.text, 'html.parser')
for keyword in soup.findAll('td'):
    keyword = keyword.find('a')
    if(keyword!=None):
        sql_keywords.append(keyword.text)
```

```python
import re

sql_keywords_counter = []

for record in sql_records_df["parsed_url"]:
    tmp = 0
    for keyword in sql_keywords:
        if(re.search(keyword,record)):
            tmp = tmp +1
    sql_keywords_counter.append(tmp)

sql_records_df["sql_keyword"] = sql_keywords_counter
```

```python
sql_records_df
```

```python
params_length = []

for record in sql_records_df["parsed_url"]:
    tmp = 0
    params = []

    for i in parse_qs(record).values():
       # print(str(i))
       # print(re.search(r"\['(\b.*)'\]",str(i)).group(1))
       # params.append(re.search(r"\['(\b.*)'\]",str(i)).group(1))
        params.append(str(i)[2:-2])
    
    for param in params:
        tmp = tmp + len((param))

    params_length.append(tmp)
  
sql_records_df["params_length"] = params_length
```

```python
print(len(sql_records_df.loc[sql_records_df["params_length"] < 10]))
sql_records_df = sql_records_df[sql_records_df["params_length"] != 2]
print(len(sql_records_df))
```

```python
sql_records_df
```

```python
indexes = []
base_event_time = 0
period = 10 # value in seconds


sql_records_df["request_volume"] = "NaN"
hostname_list = sql_records_df.hostname.unique()

for hostname in hostname_list:
    base_event_time = sql_records_df.loc[sql_records_df["hostname"] == hostname]["event_time"].to_list()[0]

    for time in sql_records_df.loc[sql_records_df["hostname"] == hostname]["event_time"].to_list():
        if(time < (float(base_event_time) + float(period))):
            indexes.append(sql_records_df.loc[(sql_records_df["hostname"] == hostname) & (sql_records_df["event_time"] == time)].index[0])  
                   
        else:
            #print(indexes)
            for index in indexes:
                sql_records_df.at[index , "request_volume"] = len(indexes)/period 
            
            indexes = []
            indexes.append(sql_records_df.loc[(sql_records_df["hostname"] == hostname) & (sql_records_df["event_time"] == time)].index[0])  
            base_event_time =  time
            #print(f'{base_event_time} --  {indexes}')
            
    for index in indexes:
        sql_records_df.at[index , "request_volume"] = len(indexes)/period
    indexes = []

   
```

```python
sql_records_df
```

```python
# * need to be escaped in regexp engine to be interpreted as character
special_chars_list = ['--', '\'' , '\"', ':', ';', '#', '/\*', '\*/', ',', '.', '\(' , '\)']

special_chars_count = []

for record in sql_records_df["parsed_url"]:
    counter_tmp = 0
    for element in special_chars_list:
        if(re.search(element, record)):
            counter_tmp = counter_tmp + 1
    special_chars_count.append(counter_tmp)

sql_records_df["special_chars"] = special_chars_count
```

```python
sql_records_df
```

```python
sql_records_df["values_to_length"] = (sql_records_df["sql_keyword"] + sql_records_df["special_chars"]) * 100 / sql_records_df["params_length"]
```

```python
sql_records_df = sql_records_df.reindex(columns=['full_url', 'parsed_url', 'hostname', 'event_time',
                                                'sql_keyword', 'params_length', 'request_volume', 
                                                'special_chars', 'values_to_length', 'injection'])
```

```python
sql_records_df.to_csv("sql_df_preprocessed.csv", index=False)
```
